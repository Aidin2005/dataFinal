```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("Preprocessed_before_standardization_and_normalization.csv")
```

```python
df['rooms'] = df['rooms'].replace({
    '6 и более': 6,
    'свободная планировка': 7
}).astype(int)
df = df.drop(df.nlargest(5, 'building_age').index)
```

```python

sns.histplot(data=df, x='m2_price', bins=30, kde=True)
plt.title('Distribution of Price per m²')
plt.xlabel('Price per m² ($)')
plt.ylabel('Number of Apartments')
plt.show()
```

```python
df['price'].plot(kind='hist', bins=30, figsize=(10,6), color='skyblue', edgecolor='black')
plt.title('Distribution of Apartment Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_price_by_district = df.groupby('district')['price'].mean().sort_values()

avg_price_by_district.plot(kind='bar', figsize=(10,6))
plt.title('Average Apartment Price by District')
plt.ylabel('Average Price ($)')
plt.xlabel('District')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# 0 = Свердловский район
# 1 = Октябрьский район
# 2 = Ленинский район
# 3 = Первомайский район
```

```python
sns.boxplot(data=df, x='rooms', y='price')
plt.title('Price Distribution by Number of Rooms')
plt.xlabel('Rooms')
plt.ylabel('Price ($)')
plt.show()
```

```python
df['building_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Building Type Distribution')
plt.ylabel('')
plt.show()
#0 = панельный
#1 = монолитный
#2 = кирпичный
```

```python
sns.scatterplot(data=df, x='square', y='price', hue='rooms')
plt.title('Price vs. Area (Colored by Room Count)')
plt.xlabel('Area (m²)')
plt.ylabel('Price ($)')
plt.legend(title='Rooms')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='is_good_floor', y='price', data=df)
plt.title('Price Distribution by Floor Type')
plt.xlabel('Is Good Floor')
plt.ylabel('Price ($)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='building_age', y='price', alpha=0.6)

plt.title('Apartment Price vs Building Age')
plt.xlabel('Building Age (years)')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```
